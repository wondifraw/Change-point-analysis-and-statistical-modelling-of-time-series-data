# Events Analysis and Validation

Detailed analysis of compiled geopolitical and economic events.

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sys.path.append('..')
from src.event_compiler import EventCompiler
```

```python
# Load and validate events
compiler = EventCompiler()
events_df = compiler.compile_major_events()
validation = compiler.validate_events_data(events_df)

print("Events Validation Results:")
for key, value in validation.items():
    print(f"{key}: {value}")

# Display events timeline
plt.figure(figsize=(14, 8))
colors = {'Geopolitical': 'red', 'Economic': 'blue', 'OPEC Decision': 'green'}
for category in events_df['category'].unique():
    data = events_df[events_df['category'] == category]
    plt.scatter(data['date'], [category]*len(data), 
               c=colors[category], s=100, alpha=0.7, label=category)

plt.title('Major Oil Market Events Timeline')
plt.xlabel('Date')
plt.ylabel('Event Category')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

events_df
```
